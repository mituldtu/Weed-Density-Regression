```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
weeddata = pd.read_excel("Weed regression final.xlsx")
```

```python
import numpy as np

from numpy.polynomial.polynomial import polyfit


x = weeddata["Density of Weed 2022"]
y = weeddata["Grain Yield 2022"]

x1 = weeddata["Density of Weed 2023"]
y1 = weeddata["Grain Yield 2023"]

b, m = polyfit(x, y, 1)

plt.figure(figsize=(4,4))
plt.plot(x, y, '.', label = '2021-22')
plt.plot(x, b + m * x, '-',label = '2021-22')
plt.legend()



b, m = polyfit(x1, y1, 1)

plt.plot(x1, y1, '.', label = '2022-23')
plt.plot(x1, b + m * x1, '--',label = '2022-23')
plt.legend()


plt.show()
```

```python
Grain Yield vs PMinor Weed Density regression model
```

```python
from sklearn import linear_model
regression_model = linear_model.LinearRegression()
regression_model.fit(X= pd.DataFrame(weeddata["Density of Weed 2022"]), y = weeddata["Grain Yield 2022"])

train_prediction = regression_model.predict(X = pd.DataFrame(weeddata["Density of Weed 2022"]))

regression_model.score(X= pd.DataFrame(weeddata["Density of Weed 2022"]), y = weeddata["Grain Yield 2022"])
```

```python
R2 of regression model between PMinor Density of Weed vs Grain Yield 2021-22
```
